## show the keyframe from orb

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pos_1 = []
pos_2 = []
pos_3 = []
pos_4 = []
cnt = 0

path = 1
slice = 0

f = open("KeyFrameTrajectory%d.txt"%path, "r")
for line in f:
    if line == "\n":
        cnt += 1
        continue
    if cnt == slice:
        data = line.split(' ')
        if path == 0:
            pos_1.append(float(data[1]))
            pos_2.append(float(data[2]))
            pos_3.append(float(data[3]))
        else:
            pos_1.append(float(data[0]))
            pos_2.append(float(data[1]))
            pos_3.append(float(data[2])) 

# path/slice
# 0 := 306 points, laptop cam
# 1 := 520 points, pi with a good trajctory "Î“"-shape
# 2 := 222 points, pi trajctory "C"-shape
# 3 := 410 points, pi good trajctory "F"-shape
# 4/0 := 247 points, pi setting
# 4/1 := 16 points, pi setting
# 4/2 := 10 points, pi setting
# 4/3 := 19 points, pi setting
# 5 := 367 points, pi setting
# 6 := 85 points, pi setting?
# 7 := 55 points, pi setting?
```

```python
pos_x= np.array(pos_1)
pos_y= np.array(pos_2)
pos_z= np.array(pos_3)



plt.plot(pos_x, pos_y,c='g')
plt.scatter(pos_x[0], pos_y[0])
plt.scatter(pos_x[-1], pos_y[-1])

plt.show()
```

```python
# Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot(pos_1, pos_2, pos_3, lw=2)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

plt.show()
```
